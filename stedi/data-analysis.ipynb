```python
import pandas as pd
import os
import json

# Path to the directory
path_to_directory = '../cleaned_data/customers/landing'

# Number of items to read
number_of_items = 10

# List all the files in the directory
files = [file for file in os.listdir(path_to_directory) if file.endswith('.json')]

# Initialize an empty list to store the DataFrames
dataframes = []

# Iterate through the files and read each JSON file (up to the specified number)
for file in files:
    df = pd.read_json(os.path.join(path_to_directory, file), lines=True)
    df['file'] = file
    
    dataframes.append(df)

# Concatenate all the DataFrames to create the final DataFrame
final_df = pd.concat(dataframes, ignore_index=True)

# Now you can work with the final DataFrame
print(final_df.head())
```
