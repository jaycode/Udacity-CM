```python
import pandas as pd
import json

PATH = "../data/customer/landing/customers-1655293787679.json"
```

```python
# Read the entire file as a single string
with open(PATH, 'r') as f:
    data = f.read()

# Split the string into individual JSON objects, removing leading/trailing whitespace
data = data.split('}{')
data[0] = data[0] + '}'
data[-1] = '{' + data[-1]
for i in range(1, len(data) - 1):
    data[i] = '{' + data[i] + '}'

# Convert each string into a Python dict, then convert the entire list into a DataFrame
df = pd.DataFrame([json.loads(item) for item in data])
df
```

```python
df.to_json('./test-output.json', orient='records', lines=True)
```
